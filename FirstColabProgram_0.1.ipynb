**Line plot**

```python
import matplotlib.pyplot as plt

x  = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y1 = [1, 3, 5, 3, 1, 3, 5, 3, 1]
y2 = [2, 4, 6, 4, 2, 4, 6, 4, 2]

plt.plot(x,y1,label='Line L')
plt.plot(x,y2,label='Line H')
plt.plot()
plt.xlabel('x-axis')
plt.ylabel('y-axix')
plt.title('Line graph example')
plt.legend()
plt.show()
```

**Bar Plot**

```python
# Look at index 4 and 6, which demonstrate overlapping cases.
import matplotlib.pyplot as plt
x1 = [1, 3, 4, 5, 6, 7, 9]
y1 = [4, 7, 2, 4, 7, 8, 3]

x2 = [2, 4, 6, 8, 10]
y2 = [5, 6, 2, 6, 2]

plt.bar(x1,y1,label='Blue Bar',color='b')
plt.bar(x2,y2,label='Green Bar',color='g')
plt.title('Bar Chart Example')
plt.legend()
plt.show()
```

**Histogram**

```python
import matplotlib.pyplot as plt
import numpy as np

# Use numpy to generate a bunch of random data in a bell curve around 5.
n=5+np.random.randn(1000) #random() 
#print(n)
m=[m for m in range(len(n))]

plt.bar(m,n)
plt.title('Raw data')
plt.show()

plt.hist(n,bins=20)
plt.title('Histogram')
plt.show()

plt.hist(n,cumulative=True,bins=20)
plt.title('Cumulative histogram')
plt.show()
```

**Scatter plot**

**Stack plot**

```python
import matplotlib.pyplot as plt
idxes = [ 1,  2,  3,  4,  5,  6,  7,  8,  9]
arr1  = [23, 40, 28, 43,  8, 44, 43, 18, 17]
arr2  = [17, 30, 22, 14, 17, 17, 29, 22, 30]
arr3  = [15, 31, 18, 22, 18, 19, 13, 32, 39]

# Adding legend for stack plots is tricky.
plt.plot([],[], color='r', label='D 1')
plt.plot([],[], color='g', label='D 2')
plt.plot([],[], color='b', label='D 3')

plt.stackplot(idxes,arr1,arr2,arr3,colors=['r','g','b'])
plt.title('Stack Plot Example')
plt.legend()
plt.show()
```

**Pie Chart**

```python
import matplotlib.pyplot as plt

labels='S1','S2','S3'
selections=[25,35,40]
colors=['c','g','y']

plt.pie(selections,labels=labels,colors=colors,
        startangle=30,
        explode=(0,0.1,0),
        autopct='%1.2F%%')
plt.axis('equal')
plt.title('Pie Chart Example')
plt.show()
```

**fill_between and alpha**

```python
import matplotlib.pyplot as plt
import numpy as np

ys=200 + np.random.randn(100)
x=[x for x in range(len(ys))]
plt.plot(x,ys,'-')
plt.fill_between(x,ys,195,where=(ys>195),facecolor='g',alpha=0.5)
plt.title('Fill_between and alpha Example')
plt.show()
```

**Subplotting using Subplot2grid**

```python
import matplotlib.pyplot as plt
import numpy as np

def random_plots():
  xs = []
  ys = []
  
  for i in range(20):
    x = i
    y = np.random.randint(10)
    
    xs.append(x)
    ys.append(y)
  
  return xs, ys

fig = plt.figure()
ax1 = plt.subplot2grid((5, 2), (0, 0), rowspan=1, colspan=2)
ax2 = plt.subplot2grid((5, 2), (1, 0), rowspan=3, colspan=2)
ax3 = plt.subplot2grid((5, 2), (4, 0), rowspan=1, colspan=1)
ax4 = plt.subplot2grid((5, 2), (4, 1), rowspan=1, colspan=1)

x, y = random_plots()
ax1.plot(x, y)

x, y = random_plots()
ax2.plot(x, y)

x, y = random_plots()
ax3.plot(x, y)

x, y = random_plots()
ax4.plot(x, y)

plt.tight_layout()
plt.show()
```
